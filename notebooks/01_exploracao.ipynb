## Produto
Estamos recebendo muitas reclamações sobre trocas de audiencistas. Quais clientes têm maior incidência de trocas ou recusas? Existem padrões por região ou tipo de audiência?

```python
df_aux = df_audiencias.copy()
# Agrupa por cliente, somando qtd_troca, qtd_declinio e contando o total de registros
df_cliente = df_aux.groupby('nome_cliente').agg(
    qtd_troca_sum=('qtd_troca', 'sum'),
    qtd_declinio_sum=('qtd_declinio', 'sum'),
    total_registros=('nome_cliente', 'size')
).sort_values(by='total_registros', ascending=False).reset_index().copy()

# calcula o percentual de troca e declinio em relação ao total
df_cliente['perc_troca'] = ((df_cliente['qtd_troca_sum'] / df_cliente['total_registros']) * 100).astype(int)
df_cliente['perc_declinio'] = ((df_cliente['qtd_declinio_sum'] / df_cliente['total_registros']) * 100).astype(int)

# Calcule um alpha entre perc_troca, perc_declinio e total_registros
df_cliente['alpha'] = (df_cliente['perc_troca'] + df_cliente['perc_declinio']) * np.power(df_cliente['total_registros'], 0.45)
# Calcula o score de gravidade de 1 a 100
alpha_min, alpha_max = df_cliente['alpha'].min(), df_cliente['alpha'].max()
df_cliente['score_gravidade'] = (((df_cliente['alpha'] - alpha_min) / (alpha_max - alpha_min)) * 100).round().astype(int)
# drop unnecessary columns
df_cliente = df_cliente.drop(columns=['alpha'])

#calcula o percentil e percentil acumulado
df_cliente['percentil_acumulado'] = ((df_cliente['total_registros'].cumsum() / df_cliente['total_registros'].sum()) * 100).astype(int)
df_cliente['percentil'] = ((df_cliente['total_registros'] / df_cliente['total_registros'].sum()) * 100).astype(int)

display(df_cliente[['nome_cliente', 'total_registros','percentil','percentil_acumulado']])
# calcule o % da quantidade de clientes que representam 80% do percentil acumulado
percentil_80 = int(df_cliente[df_cliente['percentil_acumulado'] <= 80]['nome_cliente'].size / df_cliente['nome_cliente'].size * 100)
print(f"Percentual de clientes responsável por 80% das audiêncas: {percentil_80}%")

```

```python
# quem são os maiores clientes ofensores em troca e declínio?
mean_perc_troca = df_cliente['perc_troca'].mean()
median_perc_troca = df_cliente['perc_troca'].median()
mean_perc_declinio = df_cliente['perc_declinio'].mean()
median_perc_declinio = df_cliente['perc_declinio'].median()
mean_total_registros = df_cliente['total_registros'].mean()
median_total_registros = df_cliente['total_registros'].median()

print("Média e mediana de perc_troca:      ", round(mean_perc_troca, 1), round(median_perc_troca, 1))
print("Média e mediana de perc_declinio:   ", round(mean_perc_declinio, 1), round(median_perc_declinio, 1))
print("Média e mediana de total_registros: ", round(mean_total_registros, 1), round(median_total_registros, 1))
display(df_cliente[['nome_cliente','perc_troca','perc_declinio','total_registros','score_gravidade']].sort_values(by='score_gravidade', ascending=False))
```

```python
# Dentre os top 20% clientes com maior score de gravidade, qual o percentual da quantidade de troca e declínio deles em relação ao total?
top_20_percent = df_cliente.nlargest(int(len(df_cliente) * 0.2), 'score_gravidade')
total_troca = top_20_percent['qtd_troca_sum'].sum()
total_declinio = top_20_percent['qtd_declinio_sum'].sum()
total_troca_all = df_cliente['qtd_troca_sum'].sum()
total_declinio_all = df_cliente['qtd_declinio_sum'].sum()
# quantidade de clientes que representam 20% do total
total_clientes_top_20 = top_20_percent['nome_cliente'].size
print(f"Quantidade de clientes que representam top 20% com score mais grave: {total_clientes_top_20}")
print(f"Percentual de troca dos top 20% clientes: {round((total_troca / total_troca_all) * 100, 1)}%")
print(f"Percentual de declínio dos top 20% clientes: {round((total_declinio / total_declinio_all) * 100, 1)}%")
print("\n Prescrição: Atuando em 43 clientes, que representam 20% da carteira, é possível atuar sob mais de 80% do total de trocas e declínios.")
```

## Operações
Quais fatores mais influenciam para que um parceiro falhe (revelia, ausência ou má atuação)? Conseguimos prever quais solicitações têm maior risco de falha?

```python
df = df_audiencias.copy()
# Cria coluna falha
df['falha'] = 0
df.loc[(df['houve_revelia'] > 0) | (df['houve_ausencia'] > 0) | (df['houve_ma_atuacao'] > 0), 'falha'] = 1

```

```python
# teste de correlação Univariada

# Considera o DataFrame df_audiencias previamente carregado
df = df_audiencias.copy()

# Cria a coluna 'falha'
df['falha'] = 0
df.loc[(df['houve_revelia'] > 0) | (df['houve_ausencia'] > 0) | (df['houve_ma_atuacao'] > 0), 'falha'] = 1
df2 = df[['nome_parceiro','nome_cliente','datahora_abertura_solicitacao','datahora_audiencia','prazo_para_inserir_dados','datahora_finalizacao_solicitacao','tipo','tipo_demanda','area_processo','tipo_audiencia','situacao','orgao','comarca','uf_comarca','situacao_dados','orientacoes_inseridas_cliente','falha']].copy()

print("\nCorrelação categórica (Cramér's V) com a coluna 'falha':")
def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.to_numpy().sum()
    phi2 = chi2/n
    r, k = confusion_matrix.shape
    # Correção para tabelas maiores que 2x2
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.round(np.sqrt(phi2corr / min((kcorr-1), (rcorr-1))) * 100,1)


# Cria uma função auxiliar para calcular o Cramér's V
def calc_cramers(col):
    # Cria a tabela de contingência para a coluna e a coluna 'falha'
    cont_table = pd.crosstab(df2[col], df2['falha'])
    return cramers_v(cont_table)

# Calcula o Cramér's V para todas as colunas de df2
results_df = df2.drop(columns=['falha']).columns.to_series().apply(calc_cramers).reset_index()
results_df.columns = ['coluna', 'correlação (%)']
results_df = results_df.sort_values(by='correlação (%)', ascending=False)

display(results_df)
print("Existe uma correlação moderada entre 'orgao' e 'falha'")
print("Existe uma correlação fraca a moderada entre orgao, datahora_audiencia, datahora_abertura_solicitacao, nome_parceiro, prazo_para_inserir_dados, comarca, datahora_finalizacao_solicitacao e nome_cliente e falha'")
print("É possível prever solicitações com maior risco de falha, mas não é possível prever todas as solicitações com falha")

date_cols = ['datahora_audiencia', 'datahora_abertura_solicitacao', 'prazo_para_inserir_dados', 'datahora_finalizacao_solicitacao']
cols_to_plot = df2.columns.drop(['falha'] + date_cols)
for col in cols_to_plot:
    # Agrupa por coluna e 'falha' e conta
    temp_df = df2.groupby([col, 'falha']).size().reset_index(name='count').copy()
    
    # Soma a contagem total para cada categoria do eixo x
    totals = temp_df.groupby(col)['count'].sum().reset_index()
    # Seleciona os top 15 com maior total de registros
    top15_categories = totals.sort_values('count', ascending=False).head(15)[col]
    
    # Filtra apenas os grupos que estão nos top 15
    temp_df = temp_df[temp_df[col].isin(top15_categories)]
    
    fig = px.bar(
        temp_df, 
        x=col, 
        y='count', 
        color='falha', 
        barmode='group',
        title=f"Distribuição de 'falha' por {col} (Top 15 por total registro)"
    )
    fig.show()
    
for col in date_cols:
    # Converte a coluna para datetime, se necessário
    df2[col] = pd.to_datetime(df2[col], errors='coerce')
    # Cria uma coluna com a data (sem a hora)
    df2[f"{col}_date"] = df2[col].dt.date
    # Agrupa por data e falha e conta a frequência dos registros
    date_counts = df2.groupby([f"{col}_date", 'falha']).size().reset_index(name='count')
    date_counts = date_counts.sort_values(f"{col}_date")
    
    # Normalize counts per date
    total_counts = date_counts.groupby(f"{col}_date")['count'].transform('sum')
    date_counts['norm_count'] = date_counts['count'] / total_counts * 100  # percentage
    
    fig = px.line(date_counts, x=f"{col}_date", y='norm_count', color='falha',
                  title=f"Distribuição normalizada de registros ao longo do tempo para {col}")
    fig.show()
```
